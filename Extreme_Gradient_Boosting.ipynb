```python


cols_to_drop = ['url_google_index',

                  'domain_google_index',

                  'qty_vowels_domain',

                  'server_client_domain',

                  'tld_present_params',

                  'time_response',

                  'domain_spf',

                  'qty_ip_resolved',

                  'qty_nameservers',

                  'qty_mx_servers',

                  'ttl_hostname',

                  'url_shortened']

df=df.drop(cols_to_drop, axis=1)
```

```python
rows, coloums = df.shape
print("Number of rows",rows)
print("Number of coloums",coloums)
```

```python
original_features = list(df.columns)
```

```python
original_features
```

```python
dataset_array = np.array(df)
```

```python
print(dataset_array)
```

```python
features_indices = []

attributes = ['url', 'domain', 'directory', 'file', 'params']
```

```python
new_dataset = {}
```

```python


for index, name in enumerate(original_features):

    if 'qty' in name and name.split('_')[-1] in attributes:

        features_indices.append([index, name.split('_')[-1]])

    else:

        new_dataset[name] = dataset_array[:, index]
```

```python
for index, name in enumerate(original_features): #index represents index of current and name represents the name in the current feature

  if 'qty' in name and name.split('_')[-1] in attributes: # checking 'qty' in the substring

    features_indices.append([index, name.split('_')[-1]]) # last part is from attributes

  else:

    new_dataset[name] = dataset_array[:,index] # it assigns the columns that don't meet the condition to the new dictionary
```

```python
for index, attribute in features_indices:

  if attribute == 'domain':

    if f"qty_char_{attribute}" not in new_dataset.keys():

        new_dataset[f"qty_char_{attribute}"] = np.zeros(rows)

    new_dataset[f"qty_char_{attribute}"] += dataset_array[:,index]
```

```python
df1 = pd.DataFrame(new_dataset).astype(int)
```

```python
df1[df1<-1] = -1
df1
```

```python
df1.describe()
```

```python
setup(data = df1, target ="phishing")
```

```python
best_model = compare_models()
```

```python
tuned_model = tune_model(best_model, n_iter=1, optimize='F1')
```

```python


predictions = predict_model(tuned_model, data=df1)
```
